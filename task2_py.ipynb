```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("gender_submission.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Distribution in Gender Submission Dataset")
plt.show()
```

```python
df['Survived'].value_counts(normalize=True) * 100
plt.figure(figsize=(8,4))
plt.plot(df['PassengerId'], df['Survived'], '.')
plt.title("PassengerId vs Survival")
plt.xlabel("PassengerId")
plt.ylabel("Survived")
plt.show()
```

```python
df.groupby('Survived').count()
```

```python
print("""
Key Findings:
1. Dataset contains only PassengerId and Survived.
2. No missing or duplicate values found.
3. Survival distribution is imbalanced.
4. PassengerId has no analytical significance.
5. Dataset is meant for submission, not deep EDA.
""")
```
